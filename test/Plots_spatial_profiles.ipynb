```python
from os import path
from pandas import read_csv
# import pandas as pd
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
data_dir='/Users/l304975/Ionboost/'
folder ='test'
```

```python
cols =['x0','xt','v','phi','E','ni','ne','rho','nhot','ncold','nss','difnor','dWhot','SWhot','charge']
fname = path.join(data_dir,folder,'profil.txt')
df=read_csv(fname,comment="#",header=None,sep='\s+',names=cols,index_col=False,skiprows=4)
```

```python
fig,ax1 = plt.subplots(2,2,figsize = (10,6))
plt.matplotlib.rc('xtick', labelsize=15)
plt.matplotlib.rc('ytick', labelsize=15)
LW=2

ax1[0][0].plot(df['xt']/40,df['ni'], linewidth=LW)
ax1[0][0].plot(df['xt']/40,df['ne'], linewidth=LW)
ax1[0][1].plot(df['xt']/40,df['rho'], linewidth=LW)
ax1[1][0].plot(df['xt']/40,df['v'], linewidth=LW)
ax1[1][1].plot(df['xt']/40,df['E'], linewidth=LW)

ax1[0][0].set_xlim(-2,20)
ax1[0][1].set_xlim(-2,20)
ax1[1][0].set_xlim(-2,20)
ax1[1][1].set_xlim(-2,20)

ax1[0][0].set_yscale('log')
ax1[0][0].set_ylim(1.e-4,1.2)
ax1[1][0].set_xlabel('spatial position [$\lambda_{Dh}$]', size=15)
ax1[1][1].set_xlabel('spatial position [$\lambda_{Dh}$]', size=15)
ax1[0][0].set_ylabel('densities [$n_h$]', size=15)
ax1[0][1].set_ylabel('charge separation', size=15)
ax1[1][0].set_ylabel('velocity [$c_s$]', size=15)
ax1[1][1].set_ylabel('electric field [$E_0$]', size=15)
ax1[0][0].grid(alpha=0.3)
ax1[0][1].grid(alpha=0.3)
ax1[1][1].grid(alpha=0.3)
ax1[1][0].grid(alpha=0.3)
fig.subplots_adjust(hspace=0.3, wspace=0.4)
plt.savefig(folder+'spatial_profils.png', bbox_inches = 'tight')     
```
